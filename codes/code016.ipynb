# Bayes en Machine Learning

El teorema de Bayes es muy util para saber como cuantificar la incertidumbre en situaciones donde hay conocimiento previo.

Por supuesto que el Teorema de Bayes no se queda afuera de la ciencia de datos. Los algoritmos utilizan el teorema de Bayes para procesos de optimizaci√≥n y aprendizaje.

Como sabemos en el area frecuentista trabajamos con el MLE o Maximum Likelihood Estimation [Estimaci√≥n de Maxima Verosimilitud]. Que es un esquema basado en las probabilidades basado en el punto de vista frecuentista.

Existe un an√°logo para el area Bayesiana y se llama MAP (Maximum Aposteriori) - Maximizar la Probabilidad Posterior.

### ¬øEn que consiste? 
Dado un conjunto de datos **D** ¬øcual es la probabilidad de que yo conociendo el conjunto de datos, tenga una hip√≥tesis que describa al conjunto de datos del modelo **h**? Por el teorema de Bayes sabemos que: La verosimilitud es la probabilidad de **D** dado la hip√≥tesis **h**, por la probabilidad del modelo **h** entre la probabilidad de la evidencia (Probabilidad del conjunto de datos **D**).

En el caso del MLE, nosotros nos preocup√°bamos por optimizar la verosimilitud que es el an√°logo de $P(D|h)$. Ahora en el esquema Bayesiana nosotros optimizamos la Posteriori $P(h|D)$ por eso buscamos:
$$\max P (h|D) \rightarrow P(D|h)P(h) $$
Y eso se traduce en el m√°ximo del producto de 2 probabilidades y omitiendo la probabilidad de la evidencia $P(D)$.

¬øPor que despreciamos la evidencia?
Porque sin importar cual es el modelo que estamos usando la evidencia siempre va ser la misma y para reducir el problema de calcular 3 probabilidades a solo 2, se acostumbra a hacer el $m√°ximo$ del producto de las probabilidades **Probabilidad prior x Verosimilitud**, y todos los problemas de optimizaci√≥n Bayesiana se reducen a eso. Aunque hay que poner atenci√≥n a la hip√≥tesis del modelamiento $h$ y a los par√°metros que posee.

Todo lo que diga Bayesiano en Machine Learning esta basado en este esquema, donde lo que hacemos **es hallar un m√°ximo de las probabilidades posteriores**. Que se traduce a obtener par√°metros a partir de los datos y no obtener los datos a partir de par√°metros, entonces son como esquemas inversos pero ambos se usan para optimizaci√≥n.

¬øEntonces aqu√≠ tambi√©n entran en juego los problemas de clasificaci√≥n? Si si entran. El clasificador m√°s sencillo que se puede usar a partir de Bayes es el Naive-Bayes. La palabra "Naive o ingenuo" proviene de una hip√≥tesis de **simplificaci√≥n**.

Estos clasificadores tambi√©n pueden estar basados en una filosof√≠a Bayesiana.

### ¬øC√≥mo seria?
Calcular la probabilidad $$\max P(h|D) = \max P(D|h) \times P(h) $$

Lo que se traduce en calcular la probabilidad de diferentes hip√≥tesis, es decir diferentes tipos de modelos $h$.

Lo que no se ve en la formulaci√≥n es que las probabilidades son algo complicadas de calcular porque el **Espacio Muestral** es muy grande, sin embargo si no asumo ninguna simplificaci√≥n sobre esas probabilidades decimos que tengo un **Modelo de Clasificaci√≥n de Bayes Optimo - Optimal Bayes Classifier**.

En la realidad eso no es f√°cil de hacer porque el computo que se requiere es muy elevado, as√≠ que estamos limitados por eso.

El segundo paso que se suele hacer dada esa complicaci√≥n es; que tenemos que factorizar la **Similitud** en producto de probabilidades mas sencillas, es decir ¬øcual es la probabilidad de que ciertos datos sean descritos por una hip√≥tesis de modelamiento?.
$$ P(D|h)= P(\{ x_1,x_2,x_3,...,x_n\}|h) $$
Es decir de cada data point $x_n$


Yo voy a decir que esas probabilidades se pueden escribir como probabilidades independientes, es decir se puede factorizar como un producto de probabilidades completamente independientes.
$$ P(D|h)=P(x_1|h)P(x_2|h).... $$
Cuando yo hago eso, las probabilidades que resultan son mucho mas sencillas porque yo lo que hago es iterar diferentes modelos sobre cada punto y eso reduce de manera considerable el espacio muestral, a esta hip√≥tesis de simplificaci√≥n es a la que llamamos **Naive Bayes**

Es as√≠ como Bayes tiene una participaci√≥n en un modelo de clasificaci√≥n. Pero no solamente sirven para clasificaci√≥n sino que tienen otro tipo de aplicaci√≥n.
